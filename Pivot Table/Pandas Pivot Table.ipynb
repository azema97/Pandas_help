```python
df = pd.DataFrame(
    {'Date':['2022-01-01','2022-01-01','2022-01-01','2022-01-02',
             '2022-01-02','2022-01-03','2022-01-03','2022-01-04',
             '2022-01-05','2022-01-05','2022-01-05','2022-01-05'],
    'ID':['ID 1','ID 2','ID 3','ID 1','ID 2','ID 1','ID 3','ID 3','ID 1','ID 1','ID 2','ID 3'],
    'Ratings':[9, 8, 5, 3, 2, 3, 6, 5, 4, 10, 3, 6]})
```

```python
df
```

```python
# Pivot tables
pd.pivot_table(df, columns = ['ID'], values = 'Ratings', index = ['Date'])
```

```python
# Aggregate functions
pd.pivot_table(df, columns = ['ID'], values = 'Ratings', index = ['Date'],
               aggfunc = ['sum','mean'])
```

```python
# Removing ID
pivotTable = pd.pivot_table(df, columns = ['ID'], values = 'Ratings', index = ['Date'])
pivotTable.columns.name = None
pivotTable
```

```python
# Reset Index
pivotTable = pivotTable.reset_index()
pivotTable
```
