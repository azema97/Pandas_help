# 🧪 DBSCAN
DBSCAN (Density-Based Spatial Clustering of Applications with Noise) es un algoritmo de clustering basado en densidad utilizado en aprendizaje no supervisado y análisis de datos. A diferencia de otros algoritmos de clustering como K-Means, DBSCAN no requiere que especifiques el número de clústeres de antemano. En su lugar, DBSCAN encuentra automáticamente el número de clústeres basándose en la densidad de los datos en el espacio.

La idea principal detrás de DBSCAN es que un clúster es una densa región de puntos que está separada de otras densas regiones por áreas de baja densidad. DBSCAN define clústeres como áreas contiguas de puntos que tienen una densidad mínima (número mínimo de puntos dentro de un radio específico) y están separadas por áreas donde la densidad de puntos es más baja.

En DBSCAN, hay tres tipos de puntos:

1. **Núcleo (Core) Points**: Son puntos que tienen al menos un número mínimo de puntos (MinPts) dentro de un radio ε (epsilon) alrededor de ellos. Estos puntos están en áreas densas del conjunto de datos.

2. **Puntos de Borde (Border Points)**: Son puntos que tienen menos de MinPts puntos dentro de ε, pero están dentro del radio ε de un punto central (núcleo). Estos puntos están en el límite de los clústeres.

3. **Puntos de Ruido (Noise Points)**: Son puntos que no son núcleo ni de borde. Estos puntos están en áreas de baja densidad y no pertenecen a ningún clúster.

En Python, puedes utilizar la biblioteca scikit-learn para implementar el algoritmo DBSCAN. Aquí tienes un ejemplo básico de cómo usar DBSCAN en Python con scikit-learn:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import DBSCAN
```

```python
df = pd.read_csv('data_dbscan.csv')
df.drop(['Unnamed: 0'], axis = 1, inplace = True)
df
```

```python
sns.scatterplot(x = df.X, y = df.Y)
plt.show()
```

```python
ds = DBSCAN(eps = 0.09, min_samples=2)
ds.fit(df)

# eps = radio epsilon que engloba los puntos de cada núcleo
# min_samples = Número mínimo de puntos dentro del radio de epsilon
```

```python
# Podemos obtener los labels y los centroides
labels = ds.labels_
print('Labels:\n')
print(labels)
```

```python
sns.scatterplot(x = df.X, y = df.Y, hue = labels)
plt.show()
```
