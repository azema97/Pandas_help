# 🧪 K Means (Clustering)

K Means es un algoritmo de clustering ampliamente utilizado en aprendizaje automático y minería de datos. Su objetivo es dividir un conjunto de datos en grupos o clústeres basándose en similitudes entre los datos. El "K" en K Means representa el número de clústeres que el algoritmo debe crear.

El algoritmo funciona de la siguiente manera:

1. **Inicialización**: Se eligen aleatoriamente K puntos del conjunto de datos como centroides iniciales. Un centroide es simplemente un punto que representa el centro de un clúster.

2. **Asignación de puntos al clúster más cercano**: Cada punto del conjunto de datos se asigna al clúster cuyo centroide está más cerca de ese punto en términos de distancia euclidiana.

3. **Actualización de centroides**: Una vez que todos los puntos han sido asignados a los clústeres, los centroides se recalculan como el promedio de todos los puntos asignados a ese clúster.

4. **Repetición**: Los pasos 2 y 3 se repiten hasta que los centroides de los clústeres no cambien significativamente entre iteraciones o hasta que se alcance un número predeterminado de iteraciones.

El resultado final del algoritmo es un conjunto de K clústeres, donde cada clúster contiene puntos que son similares entre sí según alguna métrica de distancia, generalmente la distancia euclidiana.

En Python, puedes usar la biblioteca scikit-learn para implementar el algoritmo K Means. Aquí tienes un ejemplo básico de cómo usar K Means en Python con scikit-learn:

```python
from sklearn.cluster import KMeans

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data_kmeans.csv')
df.drop(['Unnamed: 0'], axis = 1, inplace = True)
df
```

```python
# Mostrando los datos en un gráfico de dispersión
sns.scatterplot(x = df.X, y = df.Y)
plt.show()
```

```python
km = KMeans(n_clusters = 8, random_state = 0)

# n_clusters: el número de clusters (grupos) a formar así como el número de centroides a generar.
# random_state: Determina la generación de números aleatorios para la inicialización del centroide.
```

```python
km.fit(df)
```

```python
# Podemos obtener los labels y los centroides
labels = km.labels_
centers = km.cluster_centers_
print('Labels:\n')
print(labels)
print('\nCentroides:\n')
print(centers)
```

```python
sns.scatterplot(x = df.X, y = df.Y, hue = labels)
sns.scatterplot(x = centers[:, 0], y = centers[:,1], color = 'red', marker = "x", s = 250)
plt.show()
```
