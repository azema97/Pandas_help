```python
import pandas as pd
```

```python
df = pd.read_csv("data.csv",
                 usecols = ['Which character shot first?','Gender','Education'])
```

```python
df.head()
```

```python
df['Which character shot first?'].value_counts()
```

```python
df['Gender'].value_counts()
```

```python
df['Education'].value_counts()
```

```python
# get dummies
pd.get_dummies(df, columns = ['Which character shot first?'], drop_first = True)

# (1,0) = Han
# (0,1) = don't understand
# (0,0) = Greedo
```

```python
# Including NAN values
pd.get_dummies(df, columns = ['Which character shot first?'], drop_first = True, dummy_na = True)
```

```python
# Using a prefix
pd.get_dummies(df, columns = ['Which character shot first?'], drop_first = True, dummy_na = True, prefix = 'shot')
```

```python
# Using a prefix
pd.get_dummies(df, columns = ['Which character shot first?','Gender'], 
               drop_first = True, dummy_na = True, prefix = 'shot')
```
