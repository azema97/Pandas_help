#  Linear Regression
La regresi贸n lineal es un m茅todo estad铆stico que se utiliza para modelar la relaci贸n entre una variable dependiente (tambi茅n llamada variable respuesta) y una o m谩s variables independientes (tambi茅n llamadas variables predictoras) mediante una ecuaci贸n lineal. La ecuaci贸n de regresi贸n lineal tiene la forma:

$$ y = mx + b $$

donde:
- $ y $ es la variable dependiente,
- $ x $ es la variable independiente,
- $ m $ es la pendiente de la l铆nea,
- $ b $ es la intersecci贸n con el eje $y$ (el valor de $y$ cuando $x$ es 0).

Aqu铆 hay un ejemplo simple de regresi贸n lineal en un Jupyter Notebook utilizando Python y la biblioteca `matplotlib` para graficar y `scikit-learn` para realizar la regresi贸n:

````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
````

````python
# Generar datos de ejemplo
np.random.seed(42)
x = 2 * np.random.rand(100, 1)
y = 4 + 3 * x + np.random.randn(100, 1)
````

Antes del crear y entrenar el modelo, por lo general se crean se parten los sets de datos en `train` y `test`.
Esto se puede hacer con la librer铆a `train_test_split` y se har铆a de la siguiente manera:

```python
import pandas
from sklearn.model_selection import train_test_split

data = pd.read_csv('data.csv')

X = data.drop("variable dependiente", axis = 1)
y = data["variable dependiente"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

Como este es un ejemplo sencillo, no vamos a necesitar esto.

````python
# Crear un modelo de regresi贸n lineal
model = LinearRegression()
````

````python
# Ajustar el modelo a los datos
model.fit(x, y)
````

````python
# Hacer predicciones
x_new = np.array([[0], [2]])
y_pred = model.predict(x_new)
````

````python
# Crear un gr谩fico utilizando Seaborn
plt.figure(figsize=(6, 4))
sns.scatterplot(x=x.flatten(), y=y.flatten(), label='Datos de ejemplo')
sns.lineplot(x=x_new.flatten(), y=y_pred.flatten(), color='red', label='Regresi贸n Lineal')
plt.xlabel('Variable Independiente (x)')
plt.ylabel('Variable Dependiente (y)')
plt.legend()
plt.show()
````
