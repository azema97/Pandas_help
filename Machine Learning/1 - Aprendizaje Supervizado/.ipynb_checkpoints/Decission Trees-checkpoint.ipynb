# 🌳 Decission Trees

Un árbol de decisión es una estructura de árbol en la cual cada nodo interno representa una decisión basada en el valor de una característica, cada rama representa el resultado de esa decisión, y cada nodo hoja representa el resultado final o la etiqueta de clasificación. Los árboles de decisión son modelos de aprendizaje supervisado que se utilizan tanto para clasificación como para regresión.

La construcción de un árbol de decisión implica dividir el conjunto de datos en subconjuntos más pequeños en función de las características seleccionadas. Este proceso se realiza de manera recursiva hasta que se cumple algún criterio de parada, como alcanzar una profundidad máxima, dividir el conjunto de datos en subconjuntos suficientemente pequeños o no realizar una división si ciertos criterios de impureza no se cumplen.

En el contexto de la clasificación, cada hoja del árbol representa una clase o categoría a la que se asigna una instancia. En el caso de la regresión, la hoja representa el valor predicho para una instancia.

Un árbol de decisión es fácil de interpretar y visualizar, lo que lo convierte en una herramienta valiosa para entender el proceso de toma de decisiones del modelo.

Ejemplo de un árbol de decisión simple:

```
                      ¿X <= 2.5?
                      /        \
          Sí          /          \          No
                     /            \
            ¿Y <= 1.5?             ¿Z <= 1.0?
            /        \             /         \
        Sí           No         Sí           No
       /  \          /  \       /  \         /  \
     ClaseA ClaseB  ClaseB ClaseA ClaseA   ClaseB ClaseA
```

En este ejemplo, cada nodo interno realiza una pregunta sobre una característica del conjunto de datos, y cada hoja representa la clasificación final (en este caso, las clases A o B). El árbol evalúa las características de una instancia y sigue el camino correspondiente para llegar a una hoja que asigna la instancia a una clase específica.

En Python hay dos librerías para arboles de decisión:
1. `sklearn.tree.DecisionTreeClassifier`: para clasificación.
1. `sklearn.tree.DecisionTreeRegressor`: para regresión lineal.

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, export_text
from sklearn import tree
````

````python
# Cargar el conjunto de datos Iris como ejemplo
iris = load_iris()
X = iris.data
y = iris.target
````

````python
# Crear un modelo de árbol de decisión
model = DecisionTreeClassifier()
model.fit(X, y)
````

````python
# Visualizar el árbol de decisión
plt.figure(figsize=(12, 8))
tree.plot_tree(model, feature_names=iris.feature_names, class_names=iris.target_names, filled=True, rounded=True)
plt.show()
````

````python
# También puedes imprimir el árbol de decisión en formato de texto
tree_rules = export_text(model, feature_names=iris.feature_names)
print("Reglas del Árbol de Decisión:\n", tree_rules)
````
