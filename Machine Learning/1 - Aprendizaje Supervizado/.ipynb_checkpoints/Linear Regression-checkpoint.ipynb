# 📈 Linear Regression
La regresión lineal es un método estadístico que se utiliza para modelar la relación entre una variable dependiente (también llamada variable respuesta) y una o más variables independientes (también llamadas variables predictoras) mediante una ecuación lineal. La ecuación de regresión lineal tiene la forma:

$$ y = mx + b $$

donde:
- $ y $ es la variable dependiente,
- $ x $ es la variable independiente,
- $ m $ es la pendiente de la línea,
- $ b $ es la intersección con el eje $y$ (el valor de $y$ cuando $x$ es 0).

Aquí hay un ejemplo simple de regresión lineal en un Jupyter Notebook utilizando Python y la biblioteca `matplotlib` para graficar y `scikit-learn` para realizar la regresión:

````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
````

````python
# Generar datos de ejemplo
np.random.seed(42)
x = 2 * np.random.rand(100, 1)
y = 4 + 3 * x + np.random.randn(100, 1)
````

Antes del crear y entrenar el modelo, por lo general se crean se parten los sets de datos en `train` y `test`.
Esto se puede hacer con la librería `train_test_split` y se haría de la siguiente manera:

```python
import pandas
from sklearn.model_selection import train_test_split

data = pd.read_csv('data.csv')

X = data.drop("variable dependiente", axis = 1)
y = data["variable dependiente"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

Como este es un ejemplo sencillo, no vamos a necesitar esto.

````python
# Crear un modelo de regresión lineal
model = LinearRegression()
````

````python
# Ajustar el modelo a los datos
model.fit(x, y)
````

````python
# Hacer predicciones
x_new = np.array([[0], [2]])
y_pred = model.predict(x_new)
````

````python
# Crear un gráfico utilizando Seaborn
plt.figure(figsize=(6, 4))
sns.scatterplot(x=x.flatten(), y=y.flatten(), label='Datos de ejemplo')
sns.lineplot(x=x_new.flatten(), y=y_pred.flatten(), color='red', label='Regresión Lineal')
plt.xlabel('Variable Independiente (x)')
plt.ylabel('Variable Dependiente (y)')
plt.legend()
plt.show()
````
