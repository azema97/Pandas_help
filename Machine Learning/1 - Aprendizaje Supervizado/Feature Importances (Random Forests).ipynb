# üå≥ Feature Importances

En Random Forest de scikit-learn, `feature importances` se refiere a la **medida de la importancia relativa de cada caracter√≠stica en la predicci√≥n del resultado objetivo**. 

Esta medida se calcula durante el entrenamiento del modelo y puede ser accedida despu√©s de que el modelo ha sido ajustado.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
# Cargar el conjunto de datos Iris
iris = load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names
```

```python
# Entrenar un clasificador de Random Forest
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X, y)
```

```python
# Obtener la importancia de las caracter√≠sticas
feature_importances = rf_classifier.feature_importances_
```

```python
# Visualizar la importancia de las caracter√≠sticas
plt.figure(figsize=(5, 2.5))
plt.barh(range(len(feature_importances)), feature_importances, align='center')
plt.yticks(range(len(feature_importances)), feature_names)
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.title('Feature Importances')
plt.show()
```
