# 🧪 Melt
La función `melt()` en la biblioteca pandas de Python se utiliza para transformar un DataFrame de formato ancho (wide) en un formato largo (long), lo que puede ser útil en diversas situaciones de análisis de datos. Permite desplegar las columnas en filas y combinar múltiples columnas en una sola columna de valores y una columna de etiquetas.

**Sirve para desdinamizar una tabla, convirtiendo campos y valores de campos en dos nuevos campos, uno donde se aloja el nombre del campo anterior, y el otro donde estaba el valor de campo.**

```python
import pandas as pd

data = {
    'ID': [1, 2, 3],
    'Nombre': ['Alice', 'Bob', 'Charlie'],
    'Matematica': [85, 90, 78],
    'Sociales': [92, 88, 76],
    'Naturales': [78, 85, 92]
}

df = pd.DataFrame(data)
df
```

```python
df = pd.melt(df, id_vars=['ID', 'Nombre'], var_name='Materia', value_name='Calificación')
df
```
