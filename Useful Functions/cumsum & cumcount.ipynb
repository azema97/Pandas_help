# ✍ Cummulative Sum & Count

- La función `cumsum()` en pandas se utiliza para calcular **la suma acumulativa de valores en una serie o columna de un DataFrame**. Esta función devuelve una serie o DataFrame con la misma longitud que el original, donde cada elemento es la suma acumulativa de los elementos anteriores en la serie.
- La función `cumcount()` en pandas se utiliza para calcular la **cantidad acumulativa de ocurrencias de valores únicos en una serie o columna de un DataFrame**. Esta función es especialmente útil cuando deseas llevar un registro de la frecuencia acumulativa de valores únicos a medida que recorres una serie o columna de datos.

```python
import pandas as pd

data = {
    'Dia': ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo'],
    'Ventas': [100, 150, 80, 120, 200, 90, 110],
    'Vendedor':['Juan','Juan','Juan','Diego','Diego','Juan','Laura']
}

df = pd.DataFrame(data)
df
```

```python
# Cumsum
df['Ventas acumulativas'] = df['Ventas'].cumsum()
df
```

```python
# Cumcount: siempre hay que agregar uno por que empieza desde 0
df['Dias de ventas de Juan'] = df.groupby('Vendedor').cumcount() + 1
df
```
